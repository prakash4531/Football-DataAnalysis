```python
# Loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# read csv file (data of players)
fifa = pd.read_csv('players_20.csv')
```

```python
# Columns headings in list
fifa.columns
```

```python
# nationality 
fifa['nationality'].value_counts()
```

```python
# First ten nationality counts
fifa['nationality'].value_counts()[0:10]
```

```python
# first five
fifa['nationality'].value_counts()[0:5]
```

```python
fifa['nationality'].value_counts()[0:5].keys() # you can use 'index' in place of 'keys()''
```

```python
# making a bar-plot for nationality of players
plt.figure(figsize=(8,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),
       list(fifa['nationality'].value_counts()[0:5]),
       color = ['blue','red','yellow','green','pink'])
plt.show()
```

```python
# player salary and short name
player_salary = fifa[['short_name','wage_eur']]
```

```python
# first five players salary
player_salary.head()
```

```python
# first five highest payble players
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

```python
# making a bar plot of first five players wage_eur
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],
        list(player_salary['wage_eur'])[0:5],color = ['r','b','pink','c','g'])
plt.show()
```
